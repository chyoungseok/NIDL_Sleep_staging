# 1. New LE data
- LE004, LE010, LE011, LE014, LE019, LE020, LE021, LE023
- 위의 총 8명 사용 가능

```python
import os
import mne
import datetime
import pandas as pd
import numpy as np
```

```python
# path_subjects_hypno = 'D:\\USC\\Sleep dataset\\Samsung_data\\Staging_label_for_new_edf'
path_subjects_hypno = 'G:\\다른 컴퓨터\\내 노트북\\USC\\Sleep dataset\\Samsung_data\\Staging_label_for_new_edf'

subjects_hypno = os.listdir(path_subjects_hypno)
subjects_hypno.sort()
print(subjects_hypno)
print(f'\n-- {len(subjects_hypno)} available hypno_labels as txt file\n')

total_hypno_start = []
total_hypno_end = []

for subject in subjects_hypno:
    path_temp_hypno = os.path.join(path_subjects_hypno, subject)
    with open(path_temp_hypno, 'r') as f:
        hypno_lines = f.readlines()
        hypno_start = hypno_lines[18].strip().split()[1]
        hypno_end = hypno_lines[-1].strip().split()[1]

        print(f'[hypno] {subject[0:5]} {hypno_start} ~ {hypno_end}')
        
        total_hypno_start.append(hypno_start)
        total_hypno_end.append(hypno_end)

dic_total = {
    'hypno_start': total_hypno_start,
    'hypno_end': total_hypno_end,
    'edf_start': np.zeros(len(subjects_hypno)),
    'edf_duration': np.zeros(len(subjects_hypno)),
    'edf_end': np.zeros(len(subjects_hypno)),
    'start_mismatch': np.zeros(len(subjects_hypno)),
    'end_mismatch': np.zeros(len(subjects_hypno))
}

df_total = pd.DataFrame(dic_total, index = subjects_hypno)
df_total
```

```python
# path_subjects_edf = 'D:\\USC\\Sleep dataset\\Samsung_data\\NEW_EDF'
path_subjects_edf = 'G:\\다른 컴퓨터\\내 노트북\\USC\\Sleep dataset\\Samsung_data\\NEW_EDF'

subjects_edf = os.listdir(path_subjects_edf)
subjects_edf.sort()
print(subjects_edf)
print(f'\n-- {len(subjects_edf)} available edf files as edf file\n')

total_edf_start = []
total_edf_duration = []

print("subject   edf_start    edf_duration")
for i in range(len(subjects_edf)):
    path_temp_edf = os.path.join(path_subjects_edf, subjects_edf[i], 'Traces.edf')
    raw = mne.io.read_raw_edf(path_temp_edf, preload=True, verbose=False)
    edf_start = str(raw.info['meas_date'])[11:19]
    edf_duration = str(datetime.timedelta(seconds=len(raw.times)/raw.info['sfreq']))
#     print("subject   edf_start    edf_duration")
    print(f"{subjects_edf[i][0:5]}     {edf_start}     {edf_duration} ")
    
    total_edf_start.append(edf_start)
    total_edf_duration.append(edf_duration)
    df_total.iloc[i,2] = edf_start
    df_total.iloc[i,3] = edf_duration
```

# 2. VD dataset

```python
#path_subjects_hypno = 'D:\\USC\\Sleep dataset\\Samsung_2nd\\VD_sleep_scoring'
path_subjects_hypno = 'G:\\다른 컴퓨터\\내 노트북\\USC\\Sleep dataset\\Samsung_2nd\\VD_sleep_scoring'

subjects_hypno = os.listdir(path_subjects_hypno)
subjects_hypno.sort()
print(subjects_hypno)
print(f'\n-- {len(subjects_hypno)} available hypno_labels as txt file\n')

print("num  subject   hypno_start    hypno_end")
i = 1
for subject in subjects_hypno:
    path_temp_hypno = os.path.join(path_subjects_hypno, subject)
    with open(path_temp_hypno, 'r') as f:
        hypno_lines = f.readlines()
        test_24h = hypno_lines[17].strip().split()
        
        if len(test_24h) > 4:
            hypno_start = hypno_lines[17].strip().split()[2]
            hypno_end = hypno_lines[-1].strip().split()[2]            
        else:
            hypno_start = hypno_lines[17].strip().split()[1]
            hypno_end = hypno_lines[-1].strip().split()[1]

        if i < 10:
            print(f'0{i}   {subject[0:5]}     {hypno_start}       {hypno_end}')
        else:
            print(f'{i}   {subject[0:5]}     {hypno_start}       {hypno_end}')
        i += 1
        total_hypno_start.append(hypno_start)
        total_hypno_end.append(hypno_end)
        
        df_total.loc[subject[0:5]] = [hypno_start, hypno_end, 0, 0, 0]
```
