```python
import os
import mne
import datetime
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```python
path_subjects_hypno = 'D:\\samsung_original\\Samsung_3rd_PE\\scoring'
# path_subjects_hypno = 'G:\\다른 컴퓨터\\내 노트북\\USC\\Sleep dataset\\Samsung_2nd\\VD_sleep_scoring'

subjects_hypno = os.listdir(path_subjects_hypno)
subjects_hypno.sort()
print(subjects_hypno)
print(f'\n-- {len(subjects_hypno)} available hypno_labels as txt file\n')

total_hypno_start = []
total_hypno_end = []
total_index = []

for subject in subjects_hypno:
    path_temp_hypno = os.path.join(path_subjects_hypno, subject)
    with open(path_temp_hypno, 'r') as f:
        hypno_lines = f.readlines()
    hypno_start = hypno_lines[18].strip().split()[1]
    hypno_end = hypno_lines[-1].strip().split()[1]

    print(f'[hypno] {subject[0:5]} {hypno_start} ~ {hypno_end}')

    total_hypno_start.append(hypno_start)
    total_hypno_end.append(hypno_end)
    total_index.append(subject[0:8])    

dic_total = {
    'hypno_start': total_hypno_start,
    'hypno_end': total_hypno_end,
    'edf_start': np.zeros(len(subjects_hypno)),
    'edf_duration': np.zeros(len(subjects_hypno)),
    'edf_end': np.zeros(len(subjects_hypno)),
    'start_mismatch': np.zeros(len(subjects_hypno)),
    'end_mismatch': np.zeros(len(subjects_hypno))
}

df_total = pd.DataFrame(dic_total, index = total_index)
df_total
```

```python
path_subjects_edf = 'D:\\samsung_original\\Samsung_3rd_PE\\edf'
# path_subjects_edf = 'G:\\다른 컴퓨터\\내 노트북\\USC\\Sleep dataset\\Samsung_2nd\\VD_edf'

subjects_edf = os.listdir(path_subjects_edf)
subjects_edf.sort()
print(subjects_edf)
print(f'\n-- {len(subjects_edf)} available edf files as edf file\n')

total_edf_start = []
total_edf_duration = []

i = 0
for i in tqdm(range(len(subjects_edf)), desc='EDF load and get meas info ... '):
    path_temp_edf = os.path.join(path_subjects_edf, subjects_edf[i], 'Traces.edf')
    raw = mne.io.read_raw_edf(path_temp_edf, preload=True, verbose=False)
    
    edf_start = str(raw.info['meas_date'])[11:19]
    edf_duration = str(datetime.timedelta(seconds=len(raw.times)/raw.info['sfreq']))
    
    #     print("subject   edf_start    edf_duration")
    
    print(f"{subjects_edf[i]}     {edf_start}     {edf_duration} ")
    total_edf_start.append(edf_start)
    total_edf_duration.append(edf_duration)

    df_total.loc[total_index[i], 'edf_start'] = edf_start
    df_total.loc[total_index[i], 'edf_duration'] = edf_duration
```

```python
df_total.to_csv('meas_time_info.csv')
df_total
```
