```python
def stage_percentage(hypno):
    total_epoch_num = len(hypno)
    dic = {
        'W_per': hypno[hypno==0].size / total_epoch_num * 100,
        'REM_per': hypno[hypno==4].size / total_epoch_num * 100,
        'N1_per': hypno[hypno==1].size / total_epoch_num * 100,
        'N2_per': hypno[hypno==2].size / total_epoch_num * 100,
        'N3_per': hypno[hypno==3].size / total_epoch_num * 100        
    }
    return dic
```
